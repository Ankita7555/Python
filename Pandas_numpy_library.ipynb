```python
# Question 2 :

# Load the data
df_raw_load = pd.read_csv('adult-modified-01-12-2025.csv')

# Replace '?' with NaN (in case they weren't read as missing)
raw_dataframe = df_raw_load.replace('?', np.nan)

# Step 1: Compute the number of missing values for each attribute
#print to display in preoper format
print("="*60)
print("MISSING VALUES PER ATTRIBUTE")
print("="*60)
missing_counts_data = raw_dataframe.isna().sum().sort_values(ascending=False)
print(missing_counts_data)
print(f"\nTotal missing values: {raw_dataframe.isna().sum().sum()}\n")

# Step 2: Create a copy for cleaning
clean_data  = raw_dataframe.copy()            
# Step 3: Identify numeric columns
num_cols = ['age', 'hours-per-week', 'education-num']

# Step 4: Fill numeric missing values with mean
print("="*60)
print("FILLING NUMERIC MISSING VALUES WITH MEAN")
print("="*60)
for col in num_cols:
    clean_data[col] = clean_data[col].astype(float)
    mean_value = clean_data[col].mean()
    clean_data[col] = clean_data[col].fillna(mean_value)
    print(f"{col}: mean = {mean_value:.2f}")

print()

# Step 5: Identify categorical columns (all except numeric)
cat_cols = [col for col in clean_data.columns if col not in num_cols]

# Step 6: Drop rows with missing categorical values
rows_before = len(clean_data)
clean_data = clean_data.dropna(subset=cat_cols)
rows_after = len(clean_data)
print(f"Dropped {rows_before - rows_after} rows with missing categorical values")
print(f"Remaining rows: {rows_after}\n")

# Step 7: Remove 'education' and 'native-country' columns
clean_data = clean_data.drop(columns=['education', 'native-country'])
print(f"Removed columns: education, native-country")
print(f"Remaining columns: {list(clean_data.columns)}\n")

# Step 8: Verify no missing values remain
print("="*60)
print("VERIFICATION: MISSING VALUES AFTER CLEANING")
print("="*60)
print(clean_data.isna().sum())
print(f"\nTotal missing values: {clean_data.isna().sum().sum()}")
print(f"Any missing values? {clean_data.isna().any().any()}\n")

# Step 9: Display new mean and standard deviation for numeric attributes
print("="*60)
print("MEAN VALUES FOR NUMERIC ATTRIBUTES")
print("="*60)
print(clean_data[num_cols].mean())
print()

print("="*60)
print("STANDARD DEVIATIONS FOR NUMERIC ATTRIBUTES")
print("="*60)
print(clean_data[num_cols].std())
print()

# Display final dataframe info
print("="*60)
print("FINAL CLEAN DATAFRAME")
print("="*60)
print(f"Shape: {clean_data.shape}")
print(f"Columns: {list(clean_data.columns)}")
print("\nFirst 5 rows:")
print(clean_data.head())

adult_data_dataframe=clean_data
```

```python
#Question 8
print("="*70)
print("CONVERTING DATA TO STANDARD SPREADSHEET FORMAT")
print("="*70)

# Identify categorical columns (all columns except numeric ones)
cat_cols = [col for col in clean_data.columns if col not in num_cols]

print(f"\nNumeric columns: {num_cols}")
print(f"Categorical columns: {cat_cols}")

# Create dummy variables for categorical columns
dummies = pd.get_dummies(clean_data[cat_cols], drop_first=False)

# Combine numeric columns + dummies into one dataframe
adult_numeric = pd.concat([clean_data[num_cols].reset_index(drop=True),
                           dummies.reset_index(drop=True)], axis=1)

print(f"\nOriginal shape: {clean_data.shape}")
print(f"New shape: {adult_numeric.shape}")

print("\n" + "="*70)
print("TOP 10 ROWS OF NEW DATAFRAME")
print("="*70)
print(adult_numeric.head(10))

# Save to CSV file
adult_numeric.to_csv("adult_numeric.csv", index=False)
print("\n" + "="*70)
print("FILE SAVED")
print("="*70)
print("Saved file: adult_numeric.csv")
```

```python
#Question 10
edu = clean_data['education-num'].astype(float)

education_max = edu.max()
education_min = edu.min()


if education_max == education_min:
    edu_minmax = edu * 0.0
else:
    edu_minmax = (edu - education_min) / (education_max - education_min)



# 2) Z-score standardization for all numeric attributes together

num_cols = ['age', 'hours-per-week', 'education-num']
num_df = clean_data[num_cols].astype(float).copy()

means = num_df.mean()
stds = num_df.std(ddof=0)  

# ddof=0 -> population std; consistent for standardization


stds_replaced = stds.replace(0, np.nan)

zscore_df = (num_df - means) / stds_replaced


zscore_df = zscore_df.fillna(0.0)

# 3) Show top 10 rows of hours-per-week: original, normalized, standardized

hpw = clean_data['hours-per-week'].astype(float)
hpw_min = hpw.min()
hpw_max = hpw.max()

if hpw_max == hpw_min:
    hpw_minmax = hpw * 0.0
else:
    hpw_minmax = (hpw - hpw_min) / (hpw_max - hpw_min)

hpw_z = zscore_df['hours-per-week']

hours_compare = pd.DataFrame({
    'hours-per-week_original': hpw,
    'hours-per-week_normalized': hpw_minmax,
    'hours-per-week_standardized': hpw_z
})

hours_compare.head(10)
```
